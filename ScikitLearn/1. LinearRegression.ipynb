### 공부 시간에 따른 시험 점수

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("ScikitLearn/LinearRegressionData.csv")
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values # 처음부터 마지막 컬럼 직전까지의 데이터(독립 변수)
y = dataset.iloc[:, -1].values # 마지막 컬럼 데이터(종속 변수 - 결과)
```

```python
X, y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression() # 객체 생성
reg.fit(X, y) # 학습 (모델 생성), 데이터를 가장 잘 표현하는 선
```

```python
y_pred = reg.predict(X) # X 데이터에 대한 예측
y_pred
```

```python
plt.scatter(X, y, color="blue") # 산점도
plt.plot(X, y_pred) # 선 그래프
plt.title("Score by hours")
plt.xlabel("hours")
plt.ylabel("score")
plt.show()
```

```python
print("9시간 공부했을 때 예상 점수 : ", reg.predict([[9]])) # 2차원 배열 형식 맞춰야
```

```python
reg.coef_ # 기울기
```

```python
reg.intercept_ # y절편
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("ScikitLearn/LinearRegressionData.csv")
dataset
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) # 훈련:테스트 = 8:2
```

#### 분리된 데이터를 통한 모델링

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

### 데이터 시각화(훈련 세트)

```python
plt.scatter(X_train, y_train, color="blue") # 산점도
plt.plot(X_train, reg.predict(X_train)) # 선 그래프
plt.title("Score by hours (train data)")
plt.xlabel("hours")
plt.ylabel("score")
plt.show()
```

### 데이터 시각화 (테스트 세트)

```python
plt.scatter(X_test, y_test, color="blue") # 산점도
plt.plot(X_train, reg.predict(X_train)) # 선 그래프
plt.title("Score by hours (train data)")
plt.xlabel("hours")
plt.ylabel("score")
plt.show()
```

```python
reg.coef_
```

```python
reg.intercept_
```

### 모델 평가

```python
reg.score(X_test, y_test) # 테스트 세트를 통한 모델 평가
```
